```python
import scipy.io
#-------loading the data-----------------------
mat = scipy.io.loadmat('eA10_3CL.mat');
# Jason: According to different data type, there are different ways to imput the file to working memory.
# Ex: "pictures" as pixels; "Genome sequences" as characters ATCG; "Voices" as  function of Volume, pitch and timber...etc.
#--illustration of datafile---------------------------
#+++++++[1]form of data++++++++++++++++++++++++++++
# データはmatlabの(.mat)形式で、構造体(仮に変数名をtmpとする)
#として下記の4変数が格納されています。
#(1)tmp.x1 = train画像群(600, 64, 64, 3)   <-pixel group　
#(2)tmp.x2 = test画像群(60, 64, 64, 3)     <-pixel group　
#(3)tmp.y1 = train注釈：分類ラベル(600,1)   <-label　 
#(4)tmp.y2 = test注釈:分類ラベル(60,1)      <-label　
#
#++++++[2]picture data(x1,x2)+++++++++++++++++++++++
#1つの画像は縦横が64,64で、RGBの3チャネルです。各画像  <- 64*64 pixel points and 3 color for each point
#サイズは(64,64,3)になります。train画像群の場合は     <- for each pictures, the size is 64*64*3 (Very big)
#600枚を一つの変数x1に入れたので、変数のサイズは       <- a set of 600 pictures stands for one variance.
#(600,64,64,3)になります。
#
#+++++[3]class label data(y1,y2)+++++++++++++++++++++++  
#画像注釈の分類ラベルには、クラス番号（0,1,2）が格納されています。  <- three kinds of classes
#
#+++++[4]訓練データ変数の作成+++++++++++++++++++++++++
#訓練データの変数の作成（tmp.x1からx_trainへの変換名変更）は、   <-set the training data's value as temporary data's value
#後続セルの「画像データ(X)と注釈(Y)準備」を参照してください
```

```python
import keras     # Keras is a high-level neural networks API, which is written in Python and capable of running on top of TensorFlow, CNTK, or Theano. Here we choose TensorFlow.
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Activation
from keras.layers import Conv2D, MaxPooling2D
from keras import backend as K

#-----学習用パラメータ-------------------(learning parameter)
batch_size = 128 #256    <- Off-line machine learning
num_classes = 3  #       <- three kinds of class
epochs = 12      #6      <- training six times

#----画像サイズ設定------------------- 
img_rows, img_cols = 64, 64
input_shape = (img_rows, img_cols, 3)

#----画像データ(X)と注釈(Y)準備------------ 
x_train = mat['x1']
x_test = mat['x2']
y_train = mat['y1']
y_test = mat['y2']

#-----画像データのRGB値を[0-1]に変換---------------------- color value
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')

#----クラス番号0,1,2をバイナリ[1,0,0][0,1,0][0,0,1]に変換-------
y_test_org = y_test
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)


#--------modelの定義------------------------------------- defining the model
#folloing lines are Credit assignment path (CAP), which describes the potentially causal connections between input and output.
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),                 #deep learning involves CAP depth > 2.
                 input_shape=input_shape))
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(3, activation='softmax'))

model.compile(loss='categorical_crossentropy',#損失関数 punishment for false answer.
              optimizer='adam', #最適化手法  the most critical algorithm (function) for determing the relations.
              metrics=['accuracy']) #評価関数 to show the accuracy.

#--------modelを学習----------------------------------
history=model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test))

#--------modelの評価--------------------------------- show up the result value of learning model
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss: %0.03f' % score[0])
print('Test accuracy: %0.03f' % score[1])

#Jason:
#For the first time, I set...
#  batch_size = 256  
#  num_classes = 3       
#  epochs = 6 
#    Test loss: 0.661
#    Test accuracy: 0.767
#However for another experiment...
#  batch_size = 128  
#  num_classes = 3       
#  epochs = 12
#    Test loss: 0.611
#    Test accuracy: 0.800
#conclusion1: The accuracy increase, and loss decrease for each step.
#conclusion2: "number of epochs" seems to be more critical than "size of batch"
#conclusion3: deeper is better than wider.
#input -> hidden layer -> output
```

評価尺度Accuracyの定義 # Here using the Markdown option
$$
\begin{eqnarray}
Accuracy &=& {\frac{TP+FN}{TP+FP+TN+FN}} 
\end{eqnarray}
$$

$$
\begin{array}{|c||c|c|}
　　 & Predicted:YES & Predicted:NO \\\hline 
Actual:YES & [True~Positive(TP)] &[False~Negative(FN)] \\\
Actual:NO & [False~Positive(FP)] &
[True~Negative(TN)]
\end{array}
$$
